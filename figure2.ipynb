```python
df = pd.read_csv('fig2_data.csv',encoding = "cp932")
df.head()
```

```python
df = df.drop(columns={'決算種別'})
df = df.drop(columns={'連結基準'})
df = df.drop(columns={'企業名称'})	
df = df.drop(columns={'日経会社コード'})
```

```python
df = df.rename(columns={'決算期':'AP'})
df = df.rename(columns={'<貸借対照表（資産）>資産合計(百万円)':'assets'})
df = df.rename(columns={'<その他項目>設備投資額［累計］(百万円)':'Capital expenditures'})
df = df.rename(columns={'<その他項目>研究開発費［累計］(百万円)':'R&D'})
```

```python
df.head()
```

```python
df['fyear'] = df['AP'].str[:4]
df['fyear'] = df['fyear'].astype(int)
```

```python
df.head()
```

```python
df = df.drop(columns={'AP'})
```

```python
df.head()
```

```python
df["R&D/ASSETS"]=df["R&D"]/df["assets"]
```

```python
df.head()
```

```python
df["Capital expenditures/ASSETS"]=df["Capital expenditures"]/df["assets"]
```
